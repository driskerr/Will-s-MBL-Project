```python
#Import Libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#Pull in Website as HTML
url = 'http://www.espn.com/mlb/history/leaders/_/breakdown/season/year/2018'

page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
#Get all the rows of the table
rows = soup.find_all('tr')

#Create column names from the 1st row of the table (the header)
columns = [i.get_text() for i in rows[1].find_all('td')]
columns[0] = 'Rank' #I didn't like the blank column name
columns

#create empty dataframe with column names from the header
final_df = pd.DataFrame(columns=columns)
final_df
```

```python
url2 = 'http://www.espn.com/mlb/history/leaders/_/breakdown/season/year/2018/start/51'
url3 = 'http://www.espn.com/mlb/history/leaders/_/breakdown/season/year/2018/start/101'
url4 = 'http://www.espn.com/mlb/history/leaders/_/breakdown/season/year/2018/start/151'
url5 = 'http://www.espn.com/mlb/history/leaders/_/breakdown/season/year/2018/start/201'
url6 = 'http://www.espn.com/mlb/history/leaders/_/breakdown/season/year/2018/start/251'

urls = [url, url2, url3, url4, url5, url6]
```

```python
#for every website that we're looking at ... 
for url in urls:
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    #... get the rows
    rows = soup.find_all('tr')
    #for every row in the table after the header
    for row in rows[1:]:
        #create a dictionary that links the stat with the column name
        player_row = dict(zip(columns, [i.get_text() for i in row.find_all('td')]))
        if 'PLAYER' not in player_row.values(): #we don't want those weird mid-table headers
            #turn that dictionary into a single-rowed dataframe for that player
            df_temp = pd.DataFrame.from_dict(player_row, orient='index').transpose()
            #add that player to the final dataframe
            final_df = pd.concat([final_df,df_temp],  ignore_index=True)
        else:
            pass
```
